```python
import numpy as np
import scipy.fftpack
import matplotlib.pyplot as plt

#采样点选择1400个，因为设置的信号频率分量最高为600Hz，根据采样定理知采样频率要大于信号频率2倍，所以这里设置采样频率为1400Hz（即一秒内有1400个采样点）
x=np.linspace(0,1,1400)

#设置需要采样的信号，频率分量有180，390和600
y=7*np.sin(2*np.pi*180*x) + 1.5*np.sin(2*np.pi*390*x)+5.1*np.sin(2*np.pi*600*x)
```

```python
%matplotlib inline

fs = 1 / (x[1] - x[0])
y_data = y

length = len(y_data)

x_data = abs(scipy.fftpack.fftfreq(len(y_data), 1/fs))
yf = scipy.fftpack.fft(y_data)
yf = np.abs(yf) / length


pl_length = length
fig, ax = plt.subplots(figsize=(16,9))
ax.plot(x_data[1:pl_length], np.abs(yf[1:pl_length]))
plt.show()
```
